```python
conn=sqlite3.connect('pets_database.db')
pd.read_sql("SELECT *FROM cats;", conn)
```

```python
pd.read_sql("""
            SELECT name,age
            FROM cats
            WHERE age>=5;
            """,conn)
```

```python
#BETWEEN
pd.read_sql("""
            SELECT name,age
            FROM cats
            WHERE age BETWEEN 1 and 3;""",conn)
```

```python
#isnull
pd.read_sql("""
            SELECT name,owner_id
            FROM cats
            WHERE owner_id IS NULL;""",conn)
```
