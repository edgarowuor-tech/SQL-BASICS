```python
import sqlite3
import pandas as pd
```

```python
conn=sqlite3.connect('data.sqlite')
```

```python
cur=conn.cursor()
```

```python
#working on employees table
pd.read_sql("""SELECT * FROM employees;""", conn)
```

```python
#retrieving a subset of columns firstname and lastname
pd.read_sql("""
SELECT lastName, firstName
  FROM employees;
""", conn).head()
```

```python
#reordering the columns
pd.read_sql("""
SELECT firstName, lastName
  FROM employees;
""", conn).head()
```

```python
#use aliases (AS keyword) to change the column names in our query result:
pd.read_sql("""
SELECT firstName AS name
  FROM employees;
""", conn).head()
```

```python
#CASE EXAMPLES
pd.read_sql("""
SELECT firstName, lastName, jobTitle,
       CASE
       WHEN jobTitle = "Sales Rep" THEN "Sales Rep"
       ELSE "Not Sales Rep"
       END AS role
  FROM employees;
""", conn).head(10)
```

```python
#CASE to Make Values Human-Readable
#In the example below, we use a CASE statement with multiple WHENs in order to transform the officeCode column into an office column that uses a more meaningful name for the office:
pd.read_sql("""
SELECT firstName, lastName, officeCode,
       CASE
       WHEN officeCode = "1" THEN "San Francisco, CA"
       WHEN officeCode = "2" THEN "Boston, MA"
       WHEN officeCode = "3" THEN "New York, NY"
       WHEN officeCode = "4" THEN "Paris, France"
       WHEN officeCode = "5" THEN "Tokyo, Japan"
       END AS office
  FROM employees;
""", conn).head(10)
```

```python
#ALTERNATIVELY
pd.read_sql("""
SELECT firstName, lastName, officeCode,
       CASE officeCode
       WHEN "1" THEN "San Francisco, CA"
       WHEN "2" THEN "Boston, MA"
       WHEN "3" THEN "New York, NY"
       WHEN "4" THEN "Paris, France"
       WHEN "5" THEN "Tokyo, Japan"
       END AS office
  FROM employees;
""", conn).head(10)
```

```python
#Built-in SQL Functions for String Manipulation
#LENGTH
#If we wanted to find the length of the first names of all employees, that would look like
pd.read_sql("""
SELECT length(firstName) AS name_length
  FROM employees;
""", conn).head()
```

```python
#UPPER
#Now let's say we wanted to return all of the employee names in all caps
pd.read_sql("""
SELECT upper(firstName) AS name_in_all_caps
  FROM employees;
""", conn).head()
```

```python
#SUBSTRING
#Let's say we wanted just the first initial (first letter of the first name) for each employee:
pd.read_sql("""
SELECT substr(firstName, 1, 1) AS first_initial
  FROM employees;
""", conn).head()
```

```python
#If we wanted to add a . after each first initial, we could use the SQLite || (concatenate) operator. 
pd.read_sql("""
SELECT substr(firstName, 1, 1) || "." AS first_initial
  FROM employees;
""", conn).head()
```

```python
#COMBINING COLUMNS
#For example, below we combine the first and last name:
pd.read_sql("""
SELECT firstName || lastName AS full_name
  FROM employees;
""", conn).head()
```

```python
#ENFORCE SPACES
pd.read_sql("""
SELECT firstName || " " || lastName AS full_name
  FROM employees;
""", conn).head()
```

```python
#SWITCHING TO ORDER DETAILS TABLE
pd.read_sql("""SELECT * FROM orderDetails;""", conn)
```

```python
#ROUNDING OFF
#Let's say we wanted to round the price to the nearest dollar.
pd.read_sql("""
SELECT round(priceEach) AS rounded_price
  FROM orderDetails;
""", conn)
```

```python
#CAST
#The previous result looks ok, but it's returning floating point numbers. What if we want integers instead?
pd.read_sql("""
SELECT CAST(round(priceEach) AS INTEGER) AS rounded_price_int
  FROM orderDetails;
""", conn)
```

```python
#BASIC MATH OPERATIONS
pd.read_sql("""
SELECT priceEach * quantityOrdered AS total_price
  FROM orderDetails;
""", conn)
```

```python
#SWITCHING TO ORDERS TABLE
pd.read_sql("""SELECT * FROM orders;""", conn)
```

```python
#What if we wanted to know how many days there are between the requiredDate and the orderDate
#Use julianday funcion
pd.read_sql("""
SELECT julianday(requiredDate) - julianday(orderDate) AS days_from_order_to_required
  FROM orders;
""", conn)
```

```python
#If we wanted to select the order dates as well as dates 1 week after the order dates, that would look like this:
pd.read_sql("""
SELECT orderDate, date(orderDate, "+7 days") AS one_week_later
  FROM orders;
""", conn)
```

```python
#USING STRFTIME TO SPLIT APART DATE OR TIME VALUE INTO SUB PARTS
pd.read_sql("""
SELECT orderDate,
       strftime("%m", orderDate) AS month,
       strftime("%Y", orderDate) AS year,
       strftime("%d", orderDate) AS day
  FROM orders;
""", conn)
```

```python
#closing the database
conn.close()
```
